# **Data Preprocessing**

```python
df = df.dropna()
```

```python
df.info()
```

# **Data Visualization**

```python
sns.pairplot(df, x_vars=['displacement', 'horsepower', 'weight', 'acceleration','mpg'], y_vars=['mpg']);
```

# **Define Target Variable (y) and Feature Variables (X)**

```python
df.columns
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
x = df[['displacement', 'horsepower', 'weight','acceleration']]
```

```python
x.shape
```

```python
x
```

# **Train Test Split**

**Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

**After Standardization Mean is Zero and Standard Deviation is One**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, train_size=0.7, random_state = 2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# **Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit (x_train, y_train)
```

# **Model Evaluation**

```python
model.intercept_
```

```python
model.coef_
```

**Mileage = 23.4 - 1.05Dispalcement + error , Mileage = 23.4 - 1.68Horsepower + error, Mileage = 23.4 - 4.10Weight + error, Mileage = 23.4 - 0.11Acceleration + error**

# **Prediction**

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

# **Explaination**

1.  mean_absolute_percentage_error = 0.14 =14% of error

    therefore accuracy of our model is 100 -14 = 86%



2.  r2_score is used to evaluate the performance of the machine learning model

    here, r2_score = 0.70

    which is close to 1

    if r2 value is 1, it means our model is well trained for unseen data

    if r2 value is o, it means our model is badly trained for unseen data

    hence our machine learning model for mileage prediction is 86% accurate
